```python
import pandas as pd
from io import StringIO

filename = 'chemicals-in-cosmetics.csv'

# Open the file and read line by line
with open(filename, 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Initialize an empty list to store rows
cleaned_lines = []

# Iterate through lines and try to append to the cleaned_lines list
for i, line in enumerate(lines):
    try:
        pd.read_csv(StringIO(line))
        cleaned_lines.append(line)
    except pd.errors.ParserError:
        print(f"Skipped line {i+1} due to parsing error.")

# Create a DataFrame from the cleaned lines
df = pd.read_csv(StringIO('\n'.join(cleaned_lines)))

# Display the first few rows of the DataFrame
df.head()
```

```python
from fuzzywuzzy import fuzz
import pandas as pd
import warnings

# Assuming df_combined_updated is a pandas DataFrame with a 'Name' column
terms = df_combined_updated['Name']

# Set a threshold for similarity
threshold = 80

# Create a DataFrame for terms
terms_df = pd.DataFrame({'Name': terms})

# Create a DataFrame for pairs of terms
pairs_df = pd.DataFrame(columns=['Term1', 'Term2'])

# Check similarity between each pair of terms
for i in range(0, len(terms) - 1, 2):
    term1 = terms.iloc[i]

    if i + 1 < len(terms):
        term2 = terms.iloc[i + 1]

        # Convert terms to strings if they are not already
        term1 = str(term1)
        term2 = str(term2)

        # Check similarity using fuzzywuzzy's partial_ratio method
        similarity_score = fuzz.partial_ratio(term1, term2)

        # If similarity score is above the threshold, add terms to the pairs DataFrame
        if similarity_score > threshold:
            pairs_df = pairs_df.append({'Term1': term1, 'Term2': term2}, ignore_index=True)
            warnings.filterwarnings("ignore")

# Sort the pairs DataFrame by 'Term1' and 'Term2'
pairs_df = pairs_df.sort_values(by=['Term1', 'Term2'])

# Create DataFrames for non-duplicate terms and duplicate terms
non_duplicates_df = pd.DataFrame(columns=['ID', 'Name'])
duplicates_df = pd.DataFrame(columns=['Original_ID', 'Duplicate_ID', 'Name'])

# Iterate through pairs and add to non-duplicates and duplicates DataFrames
for i, row in pairs_df.iterrows():
    term1 = row['Term1']
    term2 = row['Term2']

    # Add term1 to the non-duplicates DataFrame if not already present
    if term1 not in non_duplicates_df['Name'].values:
        non_duplicates_df = non_duplicates_df.append({'ID': i, 'Name': term1}, ignore_index=True)

    # Add term2 to the non-duplicates DataFrame if not already present
    if term2 not in non_duplicates_df['Name'].values:
        non_duplicates_df = non_duplicates_df.append({'ID': i, 'Name': term2}, ignore_index=True)

    # Add terms to the duplicates DataFrame
    duplicates_df = duplicates_df.append({'Original_ID': i, 'Duplicate_ID': i, 'Name': term1}, ignore_index=True)
    duplicates_df = duplicates_df.append({'Original_ID': i, 'Duplicate_ID': i, 'Name': term2}, ignore_index=True)

# Sort the DataFrames by 'Name'
non_duplicates_df = non_duplicates_df.sort_values(by='Name')
duplicates_df = duplicates_df.sort_values(by='Name')

# Print the non-duplicates DataFrame
print("Non-duplicates DataFrame:")
print(non_duplicates_df)

# Print the duplicates DataFrame
print("\nDuplicates DataFrame:")
print(duplicates_df)

```

```python
from fuzzywuzzy import process, fuzz
import pandas as pd

terms = df_combined_updated['Name']

terms_df = pd.DataFrame({'Name': terms.astype(str)})
threshold = 80

non_duplicates_set = set()
duplicates_data = []

def find_similar(row, all_terms):
    term = row['Name']
    similar_term, _ = process.extractOne(term, all_terms, scorer=fuzz.ratio)
    return similar_term

terms_df['Similar_Terms'] = terms_df.apply(lambda row: find_similar(row, terms_df['Name'].values), axis=1)

for i, row in terms_df.iterrows():
    term = row['Name']
    similar_term = row['Similar_Terms']

    non_duplicates_set.add(term)
    non_duplicates_set.add(similar_term)
    duplicates_data.append({'Original_ID': term, 'Duplicate_ID': similar_term, 'Name': term})
    duplicates_data.append({'Original_ID': term, 'Duplicate_ID': similar_term, 'Name': similar_term})

non_duplicates_df = pd.DataFrame({'ID': range(len(non_duplicates_set)), 'Name': list(non_duplicates_set)}).sort_values(by='Name')
duplicates_df = pd.DataFrame(duplicates_data).sort_values(by='Name')

print("Non-duplicates DataFrame:")
print(non_duplicates_df)

print("\nDuplicates DataFrame:")
print(duplicates_df)
```
